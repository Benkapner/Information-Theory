#### Chain rule:
In the class we proved that for RVs $X \in \Xcal$ and $Y \in $\Ycal$, 
$$
\begin{align}
H(X,Y) = H(X) + H(X|Y).
\end{align}
$$
Generalize this claim to $n$ RVs $X^n = (X_1,\ldots,X_n) \in \Xcal_1 \times \ldots \times \Xcal_n$, i.e. Prove that:
$$
H(X^n) = \sum_{i=1}^n H(X_i|X^{i-1}), 
$$
(with the notation $X^{0} = \emptyset$.). You may use induction with (1) (but not have to). 



#### <span style="color: LIGHTgreen;">Answer</span>
1. Base case for $n=2$:
$$
H(X^2) = H((X_1,X_2))=\sum_{i=1}^{2}{H(X_i|X^{i-1})}=H(X_1|X^0)+H(X_2|X^1)
$$
Given that $X^{0} = \emptyset$ we obtain:
$$H(X_1,X_2)=H(X_1)+H(X_2|X_1)$$ 
This expression is the chain rule that was proven in class.  
  
2. Inductive Hypothesis:  
We assume the claim is true for $n-1$:
$$H(X^{n-1})=\sum_{i=1}^{n-1}{H(X_i|X^{i-1})}$$  
  
3. Inductive step:
$$\sum_{i=1}^{n}{H(X_i|X^{i-1})}$$
Getting out of the summation the last item which is $H(X_n|X^{n-1})$:
$$=\sum_{i=1}^{n-1}{H(X_i|X^{i-1})}+H(X_n|X^{n-1})$$
From the inductive hypothesis we know that $H(X^{n-1})=\sum_{i=1}^{n-1}{H(X_i|X^{i-1})}$:
$$H(X^{n-1})+H(X|X^{n-1})$$
Applying chain rule we obtain:
$$=H(X_n,X^{n-1})=H(X_n,X_{n-1},...,X_1)=H(X^n)$$  
Therefore, the claim has been proven by induction.

### Concavity
A function $f: \Xcal \to \Reals$ is said to be **concave** if for any $x_1,x_2 \in \Xcal$ and $\lambda \in [0,1]$, 
$$
f(\lambda x_1 + (1-\lambda) x_2) \geq \lambda f(x_1) + (1-\lambda) f(x_2). 
$$
For example, $f(x) = \sqrt{x}$ is concave.

H( \lambda p_1 + (1-\lambda) p_2 ) \geq \lambda H(p_1) +  (1-\lambda) H(p_2). 
$$
(note that $\lambda p_1 + (1-\lambda)p_2$ is a probability distribution whenever $p_1$ and $p_2$ are such)

Hint: This can be shown in multiple ways. We suggest using our "fundamental" inequality:
$$
\ln(x) \leq x-1,\qquad x>0.
$$

Let $b$ be a variable following the distribution:
$$b = \begin{cases} 
    0, & \text{with probability }{\lambda} \\

### 3. Entropy estimation and convergence laws

#### A. Consistent estimation
Suppose that you observe a sequence of words (or letters) $x^n \in \Xcal^n$. Assume that $x^n$ represents a sample of IID RVs $X^n = (X_1,\ldots,X_n)$. Define
$$
H_n(X^n) := - \sum_{x\in \Xcal} \frac{N(x|X^n)}{n} \log \frac{N(x|X^n)}{n},
$$
where $N(x|x^n)$ denotes the number of occurrences of $x$ in the sequence $x^n$ ( $N(x|x^n) = \sum_{i=1}^n \mathbf 1\{x=x_i\}$).

- Use the law of large number (applied to $N(x|X^n)$) to show that
$$
\lim_{n\to \infty} H_n(X^n) = H(X)\quad \text{(with probability 1)},
$$
so that $H_n(X^n)$ is a consistent estimator of $H(P_X)$ (https://en.wikipedia.org/wiki/Consistent_estimator). Justify your steps. You may use without proof the continuous mapping theorem (https://en.wikipedia.org/wiki/Continuous_mapping_theorem)



#### <span style="color: LIGHTgreen;">Answer</span>

Consider a sequence of words (or letters) $ x^n \in \mathcal{X}^n $, representing a sample of IID RVs $ X^n = (X_1, \ldots, X_n) $. Define the entropy estimator:

$$ H_n(X^n) := -\sum_{x\in \mathcal{X}} \frac{N(x|X^n)}{n} \log \frac{N(x|X^n)}{n} $$

where $ N(x|x^n) $ is the number of occurrences of $ x $ in the sequence $ x^n $ ($ N(x|x^n) = \sum_{i=1}^n \mathbf{1}\{x=x_i\} $).

### Law of Large Numbers (LLN)

The Law of Large Numbers (LLN) states that, under certain conditions, the sample average of a sequence of random variables converges to the expected value with probability 1. For IID random variables $X_1, X_2, \ldots, X_n$ from $P_X$, the sample mean $\bar{X}_n = \frac{1}{n} \sum_{i=1}^n X_i$ converges to the expected value $\mu$ as $n$ approaches infinity:

$$ \frac{N(x|X^n)}{n} \xrightarrow{a.s.} P(X = x) $$

### Applying LLN to $H_n(X^n)$

Applying the LLN to $N(x | X^n)$, we get:

$$ \lim_{n\to \infty} H_n(X^n) = -\sum_{x \in \mathcal{X}} \frac{N(x|X^n)}{n}log\frac{N(x|X^n)}{n} \xrightarrow{a.s.} -\sum_{x \in \mathcal{X}} P(X = x) \log P(X = x) = H(P_X) $$


### Conclusion using the Continuous Mapping Theorem

The Continuous Mapping Theorem allows us to apply a continuous function to a converging sequence of random variables. Since the entropy function is continuous, we can conclude:

$$ \lim_{n\to \infty} H_n(X^n) = H(X) \quad \text{(with probability 1)} $$

Therefore, $H_n(X^n)$ is a consistent estimator of $H(P_X)$, converging to the true entropy $H(P_X)$ with probability 1 as the sample size $n$ goes to infinity.

#### B. Simulation
- Verify the convergence of an IID sample from $P_X = Bern(p)$ to its entropy 
$$
h_2(p) = - p\log(p)-(1-p)\log(1-p)
$$ 
by illustrating the difference between $H_n(X^n)-h_2(p)$ versus $n$ for $n=1,\ldots,N$ for $N=1,000$. Repeat for $p=0.1,0.25,0.5,0.75,0.9$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate entropy h2(p)
def h2(p):
    return -p * np.log2(p) - (1 - p) * np.log2(1 - p)

# Function to simulate entropy estimation for Bernoulli distribution
def entropy_estimate(p, n):
    x_n = np.random.binomial(1, p, n)  # IID sample from Bernoulli distribution
    occurrences = np.bincount(x_n)
    probabilities = occurrences / n
    entropy_n = -np.sum([prob * np.log2(prob) if prob > 0 else 0 for prob in probabilities])
    return entropy_n

# Parameters
N = 1000
p_values = [0.1, 0.25, 0.5, 0.75, 0.9]

# Plotting
for i, p in enumerate(p_values):
    plt.figure(figsize=(12, 8))
    entropy_differences = []
    true_entropy = h2(p)

    for n in range(1, N + 1):
        entropy_n = entropy_estimate(p, n)
        entropy_differences.append(entropy_n - true_entropy)

    plt.plot(range(1, N + 1), entropy_differences, label=f'p = {p}')
    plt.axhline(0, color='black', linestyle='--', linewidth=1)
    plt.xlabel('Sample Size (n)')
    plt.ylabel('$H_n(X^n) - h_2(p)$')
    plt.title(f'Convergence of IID Sample to Entropy (p = {p})')
    plt.legend()
    plt.show()
```

#### <span style="color: LIGHTgreen;">Answer</span>
- Property 1: $h_2(p) \leq 1$ with equality if and only if $p=1/2$.

The binary entropy is defined as: $$h_2(p)=-plog(p)-(1-p)log(1-p)$$
The binary entropy depends on $p$ that ranges from {0,1}. When calculating the binary entropy at the extreme points {0,1} we obtain:  
$$h_2(0)=-0-(1)log(1)=0$$
$$h_2(1)=-(1)log(1)-0=0$$
In question 2, we proved that the entropy is a concave function, this means that it has only one maximum. The binary entropy can be rewritten as: 
$$h_2(p)=h(p)+h(1-p)$$
Therefore, the binary entropy is a sum of two entropies, which are both concave (as proved in 2). We can thus infer that the binary entropy is also a concave function and has therefore only one maximum. To identify such maximum we calculate the first derivative of the binary entropy:
$$h_2'(p)=-log(p)+log(1-p)$$
To obtain the maximum point we equal the binary entropy to 0:
$$h_2'(p)=-log(p)+log(1-p)=0$$
$$log(p)=log(1-p)$$
$$(p)=(1-p)$$
$$p={1/2}$$
Since binary entropy is concave we know that that is the only maximum point.

- Property 2: $h_2(p) = h_2(1-p)$.

The binary entropy is defined as: $$h_2(p)=-plog(p)-(1-p)log(1-p)$$
Now let's calculate $h_2(1-p)$:
$$h_2(1-p)=-(1-p)log(1-p)-[1-(1-p)]log[1-(1-p)]$$
$$h_2(1-p)=-(1-p)log(1-p)-(p)log(p)$$
Hence, $h_2(p)=h_2(1-p)$

- Property 3: $1 - 2|p-1/2| \leq h_2(p)$ with equality for $p=0,1,1/2$.

Let's start by showing that there is equality for $p=0,1,1/2$:
- for $p=0$
$$1 - 2|p-1/2| = 1 - 2(1/2) = 0 $$
$$h_2(0)=-0log(0)-(1-0)log(1-0)=0$$
- for $p=1$:
$$1 - 2|p-1/2| = 1 - 2(1/2) = 0 $$
$$h_2(1)=-1log(1)-(1-1)log(1-1)=0$$
- for $p=1/2$:
$$1 - 2|p-1/2| = 1 - 2(0) = 1 $$
$$h_2(1/2)=-(1/2)log(1/2)-(1-1/2)log(1-1/2)=1$$
We have proven equality for all three values of p.

From property 2, we proved that the binary entropy is symmetric around $1/2$ and we also previously proved that it is a concave function and that it's maximum value of 1 is obtained when $p=1/2$.
To prove the inequality, we analyze the value of the expression $$1 - 2|p-1/2|$$  
**Case p $\geq$ $1/2$**
$$1 - 2|p-1/2| = 1-2(-(p-{1}/{2}))=2p$$
**Case p $<$ $1/2$**
$$1 - 2|p-1/2| = 2-2p$$
Hence, the inequality to be proven can be rewritten for both cases as:  
**Case p $\geq$ $1/2$**  
$$2p \leq h_2(p)$$
**Case p $<$ $1/2$**
$$2-2p \leq h_2(p)$$

As mentioned before, the binary entropy is a symmetric function around $1/2$ so it suffices to prove the inequality for one case and that will prove automatically the inequality for the other case. 
Let's prove the inequality for the first case:
**Case p $\geq$ $1/2$**  
$2p \leq h_2(p)$ can be rewritten as: $$p+p \leq -plog(p)-(1-p)log(1-p)$$
We can segment the previous expression in two parts. First we take:
$$p \leq -plog(p)$$
Which can be rewritten as:
$$plog2 \leq plog(1/p)$$
and is true for $p\leq1/2$.
On the other hand, we take:
$$p\leq-(1-p)log(1-p)$$
Which can be rewritten as:
$$-plog(p)-(1-p)log(1-p)-p \geq 0$$
Hence, we have proven that $$2p \leq h_2(p)$$ and since the binary entropy is symmetric and continuous, we can conclude that $$ 1 - 2|p-1/2| \leq h_2(p)$$

```python
import numpy as np
from matplotlib import pyplot as plt

def binary_entropy(x):
    return -x * np.log2(x) - (1 - x) * np.log2(1 - x)

def vn_extractor(bitstring):
    """
    Extract the Vn values from a bitstring

    Args:
        bitstring (np.array): A bitstring
    Returns:
        (np.array): An array of Vn values
    """

    n = len(bitstring)
    # Ignore the last bit of the bit in case the bitstring length is odd
    if n % 2 == 1:
        bitstring = bitstring[:-1]

    # Reshape the bitstring into an array of pairs
    bit_pairs = bitstring.reshape(-1, 2)

    # Initialize an empty list to store the extracted bits
    extracted_bits = []
    
    # Iterate through the bit pairs
    for bit_pair in bit_pairs:
        # If the two bits are the same, discard them
        if bit_pair[0] == bit_pair[1]:
            continue
        # Otherwise, append the first bit to the extracted bits
        else:
            extracted_bits.append(bit_pair[0])
            
    return np.array(extracted_bits)

def plot_histogram_of_bitstring_values(bitstring, word_length=2):
    """
    Plot a histogram of the values of a bitstring after grouping the bits into
    words of a given length.
    """

    while len(bitstring) % word_length != 0:
        bitstring = bitstring[:-1] # Remove last bit if necessary

    vv = 2 ** (np.arange(0, word_length)) # Vector of powers of 2
    vals2 = bitstring.reshape([-1, word_length]) @ np.expand_dims(vv, 1)  # Compute values of bitstring words
    
    plt.figure()
    plt.hist(vals2, bins=np.arange(0,2**word_length+1)-.5, density=True)
    plt.title(f"Histogram of bitstring values (word length = {word_length})")
    plt.show()


def random_bits_from_biased_bitstring(given_bitstring):
    """
    Generate a random bitstring (P(X=1) = 1/2) from a biased bitstring

    Args:
        given_bitstring (np.array): A bitstring where P(X=1) = p != 1/2
    Returns:
        (np.array): A random bitstring where P(X=1) = 1/2
    """
    # Generate a random bitstring with equal probabilities for 0 and 1
    random_bitstring = np.random.choice([0, 1], size=len(given_bitstring))

    return random_bitstring

# Demonstrate the extractor by generating random bits
k=1000
p=0.15
n=1000

extracted_bits=''
while len(extracted_bits)<k:
    biased_bitstring = np.random.choice([0, 1], size=n, p=[1 - p, p])

    # Apply von Neumann's extractor
    extracted_bits = vn_extractor(biased_bitstring)
    n+=2

# Plot histograms for different word lengths testing
plot_histogram_of_bitstring_values(biased_bitstring, word_length=1)
plot_histogram_of_bitstring_values(biased_bitstring, word_length=2)
plot_histogram_of_bitstring_values(biased_bitstring, word_length=3)
plot_histogram_of_bitstring_values(biased_bitstring, word_length=4)
plot_histogram_of_bitstring_values(biased_bitstring, word_length=5)

# Display the results
print("Expected number of samples for Bernouilli with p=0.15 to generate 1000 random bits:", n)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def extract_random_bits(bitstring):
    """
    Applies the Von Neumann extractor to a given bitstring.

    This method pairs adjacent bits and returns a new bitstring 
    where each bit is the XOR of each pair.

    Args:
        bitstring (np.array): An array of bits.

    Returns:
        np.array: An array of bits with an equal probability of 0 and 1.
    """

    # Reshape the input bitstring into pairs of bits.
    bit_pairs = bitstring.reshape(-1, 2)

    # XOR each pair of bits and return the result.
    return np.bitwise_xor(bit_pairs[:, 0], bit_pairs[:, 1])

def generate_and_plot_histogram():
    """
    Generates a random bitstring and plots a histogram of bits 
    extracted using the Von Neumann method.
    """

    # Generate a random bitstring where P(X=1) is 0.15.
    original_bitstring = np.random.rand(100000) < 0.15

    # Apply the Von Neumann extractor.
    random_bits = extract_random_bits(original_bitstring)

    # Plot the histogram of the extracted bits.
    plt.hist(random_bits, bins=np.arange(-0.5, 2), density=True)
    plt.title("Histogram of Extracted Bits")
    plt.xlabel("Bit Value")
    plt.ylabel("Density")
    plt.show()

# Run the function:
generate_and_plot_histogram()
```
